```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
# Load the data from the CSV file to inspect its contents
file_path = './reports.csv'
data = pd.read_csv(file_path)

# Display the 'OPENFOAM_WORKSPACE_DIR' column fully without truncation
pd.set_option('display.max_colwidth', None)

# Display the first few rows of the data to understand its structure
data.head()
```

```python
# Function to extract Mach number from the OPENFOAM_WORKSPACE_DIR column
def extract_mach_number(path):
    match = re.search(r'Ma(\d+\.\d+)', path)
    if match:
        return float(match.group(1))
    return None

# Apply the function to extract Mach number and create a new column for Mach
data['Mach'] = data['OPENFOAM_WORKSPACE_DIR'].apply(extract_mach_number)

# Drop rows where Mach could not be extracted
data = data.dropna(subset=['Mach'])

data.columns
```

```python
# Function to extract grid refinement level (R1, R2, etc.) from the path
def extract_refinement_level(path):
    match = re.search(r'_R(\d)', path)
    if match:
        return f"R{match.group(1)}"
    return None

# Function to extract angle of attack (AoA) from the path
def extract_aoa(path):
    match = re.search(r'AoA(\d+)', path)
    if match:
        return int(match.group(1))
    return None

# Apply functions to create new columns for grid refinement and AoA
data['Grid_Refinement'] = data['OPENFOAM_WORKSPACE_DIR'].apply(extract_refinement_level)
data['AoA'] = data['OPENFOAM_WORKSPACE_DIR'].apply(extract_aoa)

# Prepare markers and colors for distinguishing grid refinement levels and AoA
markers = {'R1': 'o', 'R2': 's', 'R3': '^', 'R4': 'D', 'R5': 'P'}
colors = {0: 'blue', 8: 'green', 10: 'red'}
```

```python
# Plot with distinguished markers and colors
plt.figure(figsize=(30, 15))

# Iterate over unique combinations of AoA and Grid refinement
for (aoa, grid), group in data.groupby(['AoA', 'Grid_Refinement']):
    plt.scatter(group['Mach'], group['CA_FINAL_COEFF'], 
                label=f'AoA: {aoa}, {grid}', 
                marker=markers.get(grid, 'o'), s=500,
                color=colors.get(aoa, 'red')
)

# Add index labels to the points
    for idx in group.index:
        plt.text(group['Mach'][idx]-0.025, group['CA_FINAL_COEFF'][idx], str(idx), fontsize=20, ha='right')

# Labels and title
plt.xlabel('Mach Number',fontsize=30)
plt.ylabel('CA_FINAL_COEFF',fontsize=30)
plt.title('CA_FINAL_COEFF vs Mach Number (Distinguished by AoA and Grid Refinement)',fontsize=30)
plt.legend(title='Legend', bbox_to_anchor=(1., 1), loc='upper left', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```
