```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import bootstrap
```

```python
# Data for M = 2.3 alpha = 16 degrees
R1R2R3 = [184254, 1129954, 2263005]
CA     = [0.355, 0.3702, 0.3687]
CN     = [1.4452, 1.601, 1.643]
CM     = [9.913, 11.63, 11.6220]
# Compute center of pressure
Xcp    = [cm / cn for cm, cn in zip(CM, CN)]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(R1R2R3, CA, marker='o', label="CA")
plt.xticks(R1R2R3, ['R1', 'R2', 'R3'])
plt.xlabel("Grid level")
plt.ylabel("C_A coefficient")
plt.title("C_A as a function of grid level")
plt.legend(loc='upper right')
plt.grid(True)
plt.savefig('CA_vs_gridlevel.eps')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(R1R2R3, CN, marker='s', label="CN")
plt.xticks(R1R2R3, ['R1', 'R2', 'R3'])
plt.xlabel("Grid level")
plt.ylabel("C_N coefficient")
plt.title("CN as a function of grid level")
plt.legend(loc='upper right')
plt.grid(True)
plt.savefig('CN_vs_gridlevel.eps')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(R1R2R3, CM, marker='^', label="CM")
plt.xticks(R1R2R3, ['R1', 'R2', 'R3'])
plt.xlabel("Grid level")
plt.ylabel("C_M coefficient")
plt.title("CM as a function of gid level")
plt.legend(loc='upper right')
plt.grid(True)
plt.savefig('CM_vs_gridlevel.eps')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(R1R2R3, Xcp, marker='+', label="Xcp")
plt.xticks(R1R2R3, ['R1', 'R2', 'R3'])
plt.xlabel("Grid level")
plt.ylabel("X_cp coefficient")
plt.title("Xcp as a function of gid level")
plt.legend(loc='upper right')
plt.grid(True)
plt.savefig('Xcp_vs_gridlevel.eps')
plt.show()
```
